## Espai de la Indexació

En aquesta secció experimentarem com augmenta l'espai ocupat per la indexació i tots els seus components a mesura que agafem més casos.

Hipòtesis: l'espai ocupat creixerà linealment amb la quantitat de casos que emmagatzema.

```python
from_model = False

increment_size = 200  # Define el tamaño del incremento
results = []

# Bucle para aumentar progresivamente la sección de df_problemes
for i in range(increment_size, df_problemes.shape[0] + increment_size, increment_size):
    upper_bound = min(i, df_problemes.shape[0])
    
    section = df_problemes.iloc[:upper_bound]
    
    rs = createRS(from_model, path="", maxcasesxleaf=10, casebase=section)

    results.append(rs.indexacio.space)

# Gráfico de los resultados
plt.figure(figsize=(10, 6))
plt.plot(range(increment_size, df_problemes.shape[0] + increment_size, increment_size), results)
plt.xlabel('Quantitat de Casos')
plt.ylabel('Bytes Ocupats')
plt.title('Evolució de l\'espai en funció de la mida del DataFrame')
plt.show()
plt.savefig('./images/espai_indexacio.png')
```
